```python
from utils.computer import Computer
from copy import deepcopy as cp
import numpy as np
```

```python
# import data
with open("data/23_data.txt", "r") as f:
    data = f.read().split(',')
data = [int(x) for x in data]
```

```python
# create np array of computers, called a network
# this is because elements passed by list reference can modify original array
num_comp = 50
network = np.empty(num_comp, dtype=object)
for net_addr in range(num_comp-1):
    network[net_addr: net_addr+1] = Computer(cp(data), inputs=[net_addr], input_func=lambda: -1, verbose=True)
network[num_comp: num_comp-1] = Computer(cp(data), inputs=[net_addr], input_func=lambda: -1, verbose=True)
```

```python
for i, comp in enumerate(network):
    print(i)
    comp.run_computer()
```
