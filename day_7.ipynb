```python
from utils.computer import Computer
from itertools import permutations
```

```python
# import data
with open("data/7_data.txt", "r") as f:
    data = f.read().split(',')
data = [int(x) for x in data]
```

```python
# set to False for part a of challenge
feedback_mode = True
outputs = []

# each phase setting is only used once
if feedback_mode:
    phases = [5, 6, 7, 8, 9]
else:
    phases = [0, 1, 2, 3, 4]
phase_combos = list(permutations(phases))


for phase_combo in phase_combos:
    
    amps = [Computer(data, phase) for phase in phase_combo]
    output = 0
    stop_flag = False
    
    while not stop_flag:
        
        for amp in amps:
            output, stop_flag = amp.run_computer(inputs=[output])
            
        if not feedback_mode:
            outputs.append(output)
            break
        elif not stop_flag:
            outputs.append(output)
        else:
            break
```

```python
ans = max(outputs)
ans
```
