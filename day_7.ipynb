```python
from utils.computer import computer
from itertools import permutations
```

```python
# import data
with open("data/7_data.txt", "r") as f:
    data = f.read().split(',')
data = [int(x) for x in data]
```

```python
max_out = 0
# set to False for part a of challenge
feedback_mode = True

# each phase setting is only used once
if feedback_mode:
    phases = [5, 6, 7, 8, 9]
else:
    phases = [0, 1, 2, 3, 4]
phase_combos = list(permutations(phases))

for phase_combo in phase_combos:
    phase_1 = phase_combo[0]
    phase_2 = phase_combo[1]
    phase_3 = phase_combo[2]
    phase_4 = phase_combo[3]
    phase_5 = phase_combo[4]
    
    out_0 = 0
    stop_flag = False
    data_1 = data.copy()
    data_2 = data.copy()
    data_3 = data.copy()
    data_4 = data.copy()
    data_5 = data.copy()
    
    pointer_1 = 0
    pointer_2 = 0
    pointer_3 = 0
    pointer_4 = 0
    pointer_5 = 0
    
    # third computer output is True when the program halts
    loop_num = 0
    while not stop_flag:
        print(f"loop number: {loop_num}")
        out_1, data_1, _, pointer_1 = computer(data_1, [phase_1, out_0], pointer_1)
        out_2, data_2, _, pointer_2 = computer(data_2, [phase_2, out_1], pointer_2)
        out_3, data_3, _, pointer_3 = computer(data_3, [phase_3, out_2], pointer_3)
        out_4, data_4, _, pointer_4 = computer(data_4, [phase_4, out_3], pointer_4)
        out_5, data_5, stop_flag, pointer_5 = computer(data_5, [phase_5, out_4], pointer_5)
        print(pointer_1, pointer_2, pointer_3, pointer_4, pointer_5)
        if not stop_flag:
            output = out_5
            out_0 = out_5
        loop_num += 1
    
        if not feedback_mode:
            stop_flag = True

    if output > max_out:
        max_out = output
```
